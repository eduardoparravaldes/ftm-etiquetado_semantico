# ConsumerKafkaToKafka

Leemos los mensajes del topic de Kafka, etiquetamos y enviamos a kafka nuevamente.  
Importante: primero ejecutamos el consumer para que este preparado esperando los mensajes y luego ejecutamos el producer.    

```python
from kafka import KafkaConsumer
consumer = KafkaConsumer('etiquetar',bootstrap_servers=['localhost:9092'])

for message in consumer:
    # e.g., for unicode: `message.value.decode('utf-8')`
    #print ("%s:%d:%d: key=%s value=%s" % (message.topic, message.partition,message.offset,message.key,message.value.decode('utf8')))
    data = message.value
    print(data)
    start = int(0)
    for word in data.split():
            palabra = word.strip(")(.,\n")
            index_ini = data.index(palabra, start)
            index_end = int(index_ini+len(unicode(palabra, encoding='utf-8')))      
            print (unicode(palabra, encoding='utf-8'), " index: ", index_ini ,index_end )
            start = index_end
```
